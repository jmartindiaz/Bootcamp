1) Implementar un juego, que consista en apilar números enteros del 1 al 20, de forma aleatoria, para lo cual debe usarse una estructura de Pila. 
Luego, el usuario debe elegir un número de veces en que se va a quitar elementos de la pila, los cuales, sumados entre sí, no deben superar el valor de 50.
El usuario pierde si la suma supera ese valor. Si no lo supera, gana, pero su calificación será 10 menos el número elementos que falten quitar para todavía no superar 50.
El programa debe informar si perdió, y si ganó, con qué calificación lo hizo.

Consideraciones:<br>
a. Se puede usar la función input() para obtener una entrada de teclado.<br>
b. Se puede usar la el modulo random para obtener valores aleatorios.

```python
import random as rnd
```

Primero creamos nuestro "MAZO DE CARTAS"

```python
pila = [rand.randint(1,20) for valor in range(5)]
pila
```

```python
pila = [rnd.randint(1, 20) for _ in range(20)] #Generamos nuestro mazo de cartas
eleccion_usuario = 5 # esta variable representa la cantidad de veces que el usuario dijo que levantaria

gano = True
acumulador = 0

for vueltas in range(eleccion_usuario):
    carta = pila.pop() # mi carta sera la carta que este mas arriba, el ultimo en ingresar y el primero en salir.
    print(f'Salió la carta {carta}')
    acumulador += carta # esta carta necesitamos ponerla en un acumulador
    print(f'Por ahora las cartas suman {acumulador}')
    if acumulador >= 50: # entonces si el acumulador iguala o pasa los 50, entonces el jugador pierde y se rompe el for
        gano = False
        print('Perdiste hermano')
        break
    # por otra parte, si el loop concluye y gano==True, entonces...
if gano:
    print(f'El usuario, sacando cartas, llegó al valor: {acumulador}')
    print('Evaluando puntaje..')
    puntaje = 10 #el puntaje inicial es 10
    nueva_sacada = pila.pop() 
    print(f'Carta sacada para ver puntaje: {nueva_sacada}')

    while acumulador + nueva_sacada < 50:
        acumulador += nueva_sacada
        puntaje -= 1
        nueva_sacada = pila.pop()
        if acumulador < 50:
            print(f'Carta sacada para ver puntaje: {nueva_sacada}')
    print(f'Puntaje final: {puntaje}')
```

2) Implementar un juego donde constas de 2 jarras, de capacidad 5 y 3 litros respectivamente, y debes colocar 4 litros en la jarra de 5L.
Las opciones posibles son:
* Llenar la jarra de 3 litros
* Llenar la jarra de 5 litros
* Vaciar la jarra de 3 litros
* Vaciar la jarra de 5 litros
* Verter el contenido de la jarra de 3 litros en la de 5 litros
* Verter el contenido de la jarra de 5 litros en la de 3 litros

Procedimiento solucion del problema:
1- Llenar la jarra de 3 litros.
Jarra grande: 0/5
Jarra chica: 3/3

2- Verter contenido de jarra de 3 litros en la de 5 litros.
Jarra grande: 3/5
Jarra chica: 0/3

3- Llenar la jarra de 3 litros.
Jarra grande: 3/5
Jarra chica: 3/3

4- Verter el contenido de la jarra de 3 litros en la de 5 litros.
Jarra grande: 5/5
Jarra chica: 1/3

5- Vaciar la jarra de 5 litros.
Jarra grande: 0/5
Jarra chica: 1/3

6- Verter el contenido de la jarra de 3 litros en la de 5 litros.
Jarra grande: 1/5
Jarra chica: 0/3

7- Llenar la jarra de 3 litros.
Jarra grande: 1/5
Jarra chica: 3/3

8- Verter el contenido de la jarra de 3 litros en la de 5 litros
Jarra grande: 4/5
Jarra chica: 0/3

Para cada accion deberia crear una clase, luego ordenar las clases

```python
class Juego_Jarra():
    def __init__(self): # construyo mi clase y especifico las variables que usare
        print ("Deberas conseguir que dejar la jarra de 5 litros con 4 litros. Para eso, las unicas acciones que potras realizar seran:")
        print ("Llenar la jarra de 3 litros\nLlenar la jarra de 5 litros\nVaciar la jarra de 3 litros\nVaciar la jarra de 5 litros\nVerter el contenido de la jarra de 3 litros en la de 5 litros\nVerter el contenido de la jarra de 5 litros en la de 3 litros\n")
        print ("QUE COMIENCE EL JUEGO...")
        self.j5 = 0 # defino a jarra 5 como un acumulador
        self.j3 = 0 # defino a jarra 3 como un acumulador

    def llenar_j3(self): # defino la funcion para llenar la jarra de 3 litros
        while (self.j3 < 3):
            self.j3 +=1
            print (f'Llenando jarra de 3 litros...\n {self.j3} litros')
        if (self.j5 == 4):
            print("Has ganado el juego, felicitaciones!")
        elif (self.j3 == 3): ### Mensaje de alerta por si el usuario se equivoca
            print ('No puedes seguir llenando, la jarra ya esta llena')

    def llenar_j5(self): # defino la funcion para llenar la jarra de 5 litros
        while (self.j5 < 5):
            self.j5 +=1
            print (f'Llenando jarra de 5 litros...\n {self.j5} litros')
        if (self.j5 == 4):
            print("Has ganado el juego, felicitaciones!")    
        elif (self.j5 == 5): ### Mensaje de alerta por si el usuario se equivoca
            print ('No puedes seguir llenando, la jarra ya esta llena')

    def vaciar_j3(self): # defino la funcion para vaciar la jarra de 3 litros
        while (self.j3 > 0):
            self.j3 -= 1
            print (f'Vaciando jarra de 3 litros... \n {self.j3} litros')
        if (self.j5 == 4):
            print("Has ganado el juego, felicitaciones!")    
        elif (self.j3 == 0): ### Mensaje de alerta por si el usuario se equivoca
            print ('No puedes vaciar, la jarra ya esta vacia')

    def vaciar_j5(self): # defino la funcion para vaciar la jarra de 5 litros
        while (self.j5 > 0):
            self.j5 -= 1
            print (f'Vaciando jarra de 5 litros... \n {self.j5} litros')
        if (self.j5 == 4):
            print("Has ganado el juego, felicitaciones!")
        elif (self.j5 == 0): ### Mensaje de alerta por si el usuario se equivoca
            print ('No puedes vaciar, la jarra ya esta vacia')

    def verter_j3_a_j5(self): # defino la funcion para verter el contenido de la jarra de 3L en la de 5L
        while (self.j3 > 0 and self.j5 < 5):
            self.j3 -= 1
            self.j5 += 1
        if (self.j5 == 4):
            print("Has ganado el juego, felicitaciones!")
        elif (self.j3 == 0):
            print ('Ya no puedes seguir vertiendo, la jarra de 3L ya esta vacia')
        elif (self.j5 == 5):
            print ('Ya no puedes seguir vertiendo, la jarra de 5L ya esta llena')
                
    def verter_j5_a_j3(self): # defino la funcion para verter el contenido de la jarra de 5L en la de 3L
        while (self.j5 > 0 and self.j3 < 3):
            self.j5 -= 1
            self.j3 += 1
        if (self.j5 == 4):
            print("Has ganado el juego, felicitaciones!")
        elif (self.j5 == 0):
            print ('Ya no puedes seguir vertiendo, la jarra de 5L ya esta vacia')
        elif (self.j3 == 3):
            print ('Ya no puedes seguir vertiendo, la jarra de 3L ya esta llena')
        

    def consultar (self): # defino la funcion para consultar los litros que contiene de cada jarra
        print (f'La jarra de 5 litros posee {self.j5} litros')
        print (f'La jarra de 3 litros posee {self.j3} litros')

```

```python
inicio = Juego_Jarra()
inicio.consultar()
print('\n 1- Llenar jarra de 3 litros \n')
inicio.llenar_j3()
inicio.consultar()
print('\n 2- Verter contenido de jarra de 3 litros en la de 5 litros. \n')
inicio.verter_j3_a_j5()
inicio.consultar()
print('\n 3- Llenar la jarra de 3 litros. \n')
inicio.llenar_j3()
inicio.consultar()
print('\n 4- Verter el contenido de la jarra de 3 litros en la de 5 litros. \n')
inicio.verter_j3_a_j5()
inicio.consultar()
print('\n 5- Vaciar la jarra de 5 litros. \n')
inicio.vaciar_j5()
inicio.consultar()
print('\n 6- Verter el contenido de la jarra de 3 litros en la de 5 litros. \n')
inicio.verter_j3_a_j5()
inicio.consultar()
print('\n 7- Llenar la jarra de 3 litros. \n')
inicio.llenar_j3()
inicio.consultar()
print('\n 8- Verter el contenido de la jarra de 3 litros en la de 5 litros \n')
inicio.verter_j3_a_j5()

```
