# Como a través de Excel no pude transformar a .csv correctamente ni tampoco cambiar el unicode, vamos a hacerlo desde Pandas:

```python
import pandas as pd
import xlrd #libreria para leer todo tipo de archivos excel: xlsx, xls, etc.
```

Los archivos .xlsx a transformar en .csv seran:
- CanalDeVenta
- Empleados
- PRODUCTOS
- Proveedores

Primero transformamos el Excel 'CanalDeVenta.xlsx'

```python
df = pd.read_excel(r'CanalDeVenta.xlsx')
df
```

En la documentacion de Pandas, el atributo 'encoding' dice lo siguiente:

encoding : string, optional
A string representing the encoding to use in the output file, defaults to ‘ascii’ on Python 2 and ‘utf-8’ on Python 3.

```python
df.to_csv('Nuevos csv\CanalDeVenta.csv', index=False, encoding='latin1') 
# en realidad el encoding utf-8 ya lo coloca por defecto pero
# no esta de mas para recordarlo.
```

En segundo lugar, transformamos el archivo 'Empleados.xlsx'

```python
df = pd.read_excel(r'Empleados.xls') # aqui el formato es .xls , es un formato de excel mas antiguo supongo
df
```

```python
df.to_csv('Nuevos csv\Empleados.csv', index=False, encoding='latin1') 
# en realidad el encoding utf-8 ya lo coloca por defecto pero
# no esta de mas para recordarlo.
```

Ahora transformamos el archivo 'PRODUCTOS.xlsx'

```python
df = pd.read_excel(r'PRODUCTOS.xlsx')
df
```

```python
df.to_csv('Nuevos csv\Productos.csv', index=False, encoding='latin1') 
```

Por ultimo, transformamos el archivo 'Proveedores.xlsx'

```python
df = pd.read_excel(r'Proveedores.xlsx')
df
```

```python
df.to_csv('Nuevos csv\Proveedores.csv', index=False, encoding='latin1') 
```
