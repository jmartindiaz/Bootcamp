```python
def obtener_coordenadas(punto):
    punto = punto.replace('POINT (', '')
    punto = punto.replace(')','')
    punto = punto.split() 

    longitude = float(punto[0])
    latitude = float(punto[1])

    return longitude, latitude
```

```python
coordenada = df.loc[0,'WKT']

localizacion = obtener_coordenadas(coordenada)
localizacion

longitude = localizacion[0]
latitude = localizacion[1]
```

Finalmente aplicamos todo esto al DataFrame entero

```python
# para  no romper los datos originales, creo una copia

datos = df.copy()

datos['longitude'] = datos['WKT'].apply(lambda x: obtener_coordenadas(x)[0])
```

```python
datos.head()
```

```python
datos['latitude'] = datos['WKT'].apply(lambda x: obtener_coordenadas(x)[1])
### datos['latitude'] me crea una nueva columna y me la agrega al final.
```

```python
datos.head()
```

```python
## seleccionar las columnas que nos piden

datos_solicitados = datos[['longitude','latitude','DOM_NORMA','NOMBRE']] ## doble corchete = entramos por label (nombre)


datos_solicitados = datos_solicitados.rename(columns={'DOM_NORMA':'name','NOMBRE':'label'})
datos_solicitados.to_csv('Datos_Solicitados_Solucion')
```
